```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import IFrame, YouTubeVideo
```

```python
df = pd.read_csv('import_data/ibmSalesProductsExample.csv')
```

```python
df.info()
```

```python
#df[df['Year']==2012].groupby([])[].agg([]).sort_values(by=[],ascending=[])  ----- default command

df[df['Year']==2012].groupby(['Product line'])['Revenue'].agg(['mean']) #.sort_values(by=[],ascending=[])
```

```python
df[df['Year']==2012].groupby(['Product line'])['Revenue'].agg(['mean']).info()
```

```python
df[df['Year']==2012].groupby(['Product line'])['Revenue','Gross margin'].agg(['mean','count']).sort_values(by=[('Revenue', 'mean')],ascending=[1])
```

# Dictionay --> {key1:value1,key2:value2}

```python
df[df['Year']==2012].groupby(['Product line'])['Revenue','Gross margin'].agg({'Revenue':['mean','max'],'Gross margin':['min','count']}).sort_values(by=[('Revenue', 'max')],ascending=[0])
```

```python
df[df['Year']==2012].groupby(['Product line','Product type'])['Revenue','Gross margin'].agg({'Revenue':['mean','max'],'Gross margin':['min','count']}).sort_values(by=[('Revenue', 'max')],ascending=[0])
```
