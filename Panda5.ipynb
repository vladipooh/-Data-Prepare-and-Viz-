```python
df.pivot_table(index=['manufacturer'],columns=['year'],values=['cty'],aggfunc=[np.mean])
```

```python
df.pivot_table(index=['manufacturer'],columns=['year'],values=['cty'],aggfunc=[np.mean]).round(2)
```

```python
df.pivot_table(index=['manufacturer'],columns=['year'],values=['cty','hwy'],aggfunc=[np.mean]).round(2).T
```

```python
df.pivot_table(index=['manufacturer','trans'],columns=['year'],values=['cty','hwy'],aggfunc=[np.mean]).round(2).T
```

```python
#Pivot table - Crosstab
pd.crosstab([df['manufacturer']],[df['class']])
```

```python
pd.crosstab([df['manufacturer']],[df['class']],margins=True)
```

```python
#Normalise = all, index, columns

pd.crosstab([df['manufacturer']],[df['class']],margins=True,normalize='all')
```

```python
(pd.crosstab([df['manufacturer']],[df['class']],margins=True,normalize='all')).round(3)
```

```python
(pd.crosstab([df['manufacturer']],[df['class']],normalize='all')).round(3).plot(kind='bar')
```

```python
(pd.crosstab([df['manufacturer']],[df['class']],normalize='all')).round(3).plot(kind='barh')
```
