```python
#https://docs.google.com/spreadsheets/d/1VE9jIHirEueZKm1QKUTbKZQAb7gyQW5BQJFoIBGUVJc/edit#gid=1326993825
url = 'https://docs.google.com/spreadsheets/d/1VE9jIHirEueZKm1QKUTbKZQAb7gyQW5BQJFoIBGUVJc/export?gid=1326993825&format=csv'
dffb = pd.read_csv(url)
```

```python
dffb[dffb['L']>10].plot(kind='bar')
```

```python
dffb[(dffb['L']>dffb['W']) & (dffb['D']>dffb['W'])].sort_values(by=['L'],ascending=0).plot(kind='bar')
```

```python
# figsize=(x,y)

dffb[(dffb['L']>dffb['W']) & (dffb['D']>dffb['W'])].sort_values(by=['L'],ascending=0).plot(kind='bar',figsize=(10,5),title='Football Stat',fontsize=15,color=['g','b','r'])
```
