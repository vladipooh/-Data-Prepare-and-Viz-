```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
# %config InlineBackend.figure_format= 'retina'
```

```python
#https://docs.google.com/spreadsheets/d/1VE9jIHirEueZKm1QKUTbKZQAb7gyQW5BQJFoIBGUVJc/edit#gid=1326993825
url = 'https://docs.google.com/spreadsheets/d/1VE9jIHirEueZKm1QKUTbKZQAb7gyQW5BQJFoIBGUVJc/export?gid=1326993825&format=csv'
dffb = pd.read_csv(url)
```

```python
dffb.set_index('Team',inplace=True)
```

```python
# figsize=(x,y)

dffb[(dffb['L']>dffb['W']) & (dffb['D']>dffb['W'])].sort_values(by=['L'],ascending=0).plot(kind='bar',
                                                                                           figsize=(10,5),
                                                                                           title='Football Stat',
                                                                                           fontsize=15,
                                                                                           color=['g','b','r'])
```

```python
dffb[0:6:1].plot(kind='bar',
                       figsize=(10,5),
                       title='Football Stat',
                       fontsize=15,
                       color=['g','b','r'])
```

```python
dffb[0:6:1][['W','L']].plot(kind='bar',
                                      figsize=(10,5),
                                      title='Football Stat',
                                      fontsize=15,
                                      color=['g','pink'])
```

```python
dffb.loc[['Manchester Utd','Liverpool']][['W','L']].plot(kind='bar',
                                      figsize=(10,5),
                                      title='Football Stat',
                                      fontsize=15,
                                      color=['g','pink'])
```

```python
dffb.loc[['Manchester Utd','Liverpool']][['L','W']].plot(kind='bar',
                                      figsize=(10,5),
                                      title='Football Stat',
                                      fontsize=15,
                                      color=['g','pink'])
```

```python
team = ['Manchester Utd','Liverpool']
see = ['W','L']
colors = ['g','pink']
ax = dffb.loc[team][see].plot(kind='bar',
                                      figsize=(10,5),
                                      title='Football Stat',
                                      fontsize=15,
                                      color=colors)
ax.set_xlabel('Good Team', fontsize=20)
ax.set_ylabel('No of Matches', fontsize=20)
ax.axhline(19,linestyle='--',linewidth=1,color='maroon')
```

```python
team = ['Manchester Utd','Liverpool']
see = ['W','L']
colors = ['g','pink']
ax = dffb.loc[team][see].plot(kind='bar',
                                      figsize=(10,5),
                                      title='Football Stat',
                                      fontsize=15,
                                      color=colors)
ax.set_xlabel('Good Team', fontsize=20)
ax.set_ylabel('No of Matches', fontsize=20)
ax.axhline(dffb['W'].mean(),linestyle='--',linewidth=1,color='maroon')
```

```python
team = ['Manchester Utd','Liverpool']
see = ['W','L']
colors = ['g','pink']
ax = dffb.loc[team][see].plot(kind='barh',
                                      figsize=(10,5),
                                      title='Football Stat',
                                      fontsize=15,
                                      color=colors)
ax.set_xlabel('No of Matches', fontsize=20)
ax.set_ylabel('Good Team', fontsize=20)
ax.axvline(dffb['W'].mean(),linestyle='--',linewidth=1,color='maroon')
```

```python
team = ['Manchester Utd','Liverpool']
see = ['W','L']
colors = ['g','pink']
ax = dffb.loc[team][see].plot(kind='bar',
                                      figsize=(10,5),
                                      title='Football Stat',
                                      fontsize=15,
                                      color=colors)

ax.set_xlabel('Good Team', fontsize=20)
ax.set_ylabel('No of Matches', fontsize=20)
ax.axhline(dffb['W'].mean(),linestyle='--',linewidth=1,color='maroon')

for i in ax.patches:
    ax.text(i.get_x()+(0.1),i.get_height()+(0.2),str(i.get_height()),fontsize=15,color='olive')
```
